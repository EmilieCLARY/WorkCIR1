<p><span style="text-decoration: underline;">EXERCICE 3<br /></span></p>
<p>L'&eacute;quation diff&eacute;rentielle</p>
<p>$$x''(t)+fx'(t)+\frac{k}{m}x(t)=0$$</p>
<p>correspond au mouvement d'une masse $m$ attach&eacute;e &agrave; un ressort de constante $k>0$ et subissant une force de  frottement inversement proportionnelle &agrave; sa vitesse multipli&eacute;e par une constante $f\geq 0$.<span style="text-decoration: underline;"><br /></span></p>
<ol>
<li>R&eacute;soudre et tracer&nbsp;sur l'intervalle [-1,10]&nbsp; les solutions pour diff&eacute;rentes valeurs de $f$ : $f=0$, $f=1$ et $f=3$ (et prendre $k=9, m=4$), avec les conditions initiales :</li>
<p>$$x(0)=1 \ \text{ et } \ x'(0)=0.$$</p>
<li>D&eacute;crire le mouvement de la masse obtenu autour de l'origine $x=0$ pour ces 3 valeurs de $f$. Expliquer ces observations d'un point de vue math&eacute;matique</li>
</ol>

```python
reset()
t=var('t')
x=function('x')(t)
xx=diff(x,t)
xx2=diff(x,t,2)
i = desolve(0==xx2(t)+(9/4)*x(t),x, ics=[0,1,0])
show(i)
j = desolve(0==xx2(t)+1*xx(t)+(9/4)*x(t),x, ics=[0,1,0])
l = desolve(0==xx2(t)+3*xx(t)+(9/4)*x(t),x, ics=[0,1,0])
plot(i, (-1, 10), color="blue") + plot(j, (-10, 10), color="red") +  plot(l, (-1, 10), color="green")
```
